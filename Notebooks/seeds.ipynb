```python
def process_cards_in_batches(cards, batch_size=300):
    token_ids = []
    total_cards = len(cards)
    
    for beg in range(0, total_cards, batch_size):
        end = min(beg + batch_size, total_cards)
        for card in cards[beg:end]:
            token_id = create_token(card)
            token_ids.append(token_id)
    
    return token_ids

# Example usage
cards = [f"card_{i}" for i in range(1500, 5000)]  # Simulate an array of card identifiers
token_ids = process_cards_in_batches(cards)

print(len(token_ids))  # Should print 5000
```
